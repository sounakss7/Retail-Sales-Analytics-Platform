```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys

sys.path.append('../src')
from retail_analytics.pipeline import run_pipeline


df, report = run_pipeline('../data/raw/superstore.csv', '../data/processed/superstore_clean.csv')
print(report)
df.head()
```

```python
monthly_sales = df.groupby(['order_year', 'order_month_name'])['sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values(['order_year', 'order_month_name'])

plt.figure(figsize=(10, 4))
sns.lineplot(data=monthly_sales, x='order_month_name', y='sales', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.tight_layout()
plt.show()
```

```python
region_summary = df.groupby('region').agg(total_sales=('sales', 'sum'), total_profit=('profit', 'sum'), order_count=('order_id', 'nunique')).reset_index()
region_summary = region_summary.sort_values('total_sales', ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(data=region_summary, x='region', y='total_sales', palette='viridis')
plt.title('Sales by Region')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
category_summary = df.groupby('category').agg(total_sales=('sales', 'sum'), total_profit=('profit', 'sum')).reset_index()

plt.figure(figsize=(8, 4))
sns.barplot(data=category_summary, x='category', y='total_profit', palette='magma')
plt.title('Profit by Category')
plt.tight_layout()
plt.show()
```
